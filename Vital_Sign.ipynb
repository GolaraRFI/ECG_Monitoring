```python
# IMPORTS
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
vidcap = cv2.VideoCapture('red_video.mp4')
frames = []
success,image = vidcap.read()

while success:
  frames.append(image)
  success,image = vidcap.read()
  # print('Read a new frame: ', success)
```

```python
#extract red channel
signal = []
for frame in frames:
  red_channel = frame[:,:,2]
  mean = np.average(red_channel)
  signal.append(mean)
```

```python
plt.rcParams["figure.figsize"] = (8,5.5)
plt.plot(signal)
plt.title('HR')
plt.grid()
plt.show()
```
